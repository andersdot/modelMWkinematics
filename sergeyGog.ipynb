```python
from astropy.io import fits
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = fits.open('galaxia_0.fits')
```

```python
data[1].data.dtype.names
```

```python
plt.scatter(data[1].data['sdss_g'] - data[1].data['sdss_i'], data[1].data['sdss_i'], s=1, lw=0)
plt.gca().invert_yaxis()
plt.show()
```

```python
plt.hist(1./(data[1].data['rad']), bins=25, log=True, histtype='step', label='truth')
plt.hist(data[1].data['plx_meas'], bins=25, log=True, histtype='step', label='measured')
plt.legend()
plt.savefig('galaxia_0_rad_plx_meas.png')
```
